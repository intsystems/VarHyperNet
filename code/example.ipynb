```python
import matplotlib.pylab as plt
import torch
import numpy as np
%matplotlib inline
```

```python
import seaborn as sns
sns.set('paper', style='white', font_scale=2)
plt.rcParams['figure.figsize'] = (6, 4.5)
plt.rcParams['font.size'] = 20
```

```python
var = torch.distributions.Normal(0.0, 1e-1)

x = []
y = []
y2 = []
for l in np.logspace(-3, 3, 100):
    prior = torch.distributions.Normal(0, np.sqrt(1/l))
    d = torch.distributions.kl_divergence(var, prior)    
    x.append(l)
    y.append(1.0 - d)
    prior = torch.distributions.Normal(0, 1)
    d2 = torch.distributions.kl_divergence(var, prior) * l    
    y2.append(1.0 - d2)
plt.plot(x,y, lw=3)
plt.xlabel('$\lambda$', fontsize=20)
plt.ylabel('$\mathfrak{L}_2$', fontsize=20)

#plt.plot(x,y2)    
plt.savefig('ex2.png',  bbox_inches = 'tight')
```

```python
plt.plot(x,y2, lw=3)  
plt.xlabel('$\lambda$', fontsize=20)
plt.ylabel('$\mathfrak{L}_1$', fontsize=20)
plt.savefig('ex1.png',  bbox_inches = 'tight')
```
